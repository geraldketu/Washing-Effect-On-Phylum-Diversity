```python
from scipy.stats import pearsonr
import pandas as pd
```

```python
data=pd.read_csv("clean_phylum_washFreq_BB.csv")
X=data['Wash_Freq of BB (# times per week)']
Y=data['NumPhylum']
```

```python
r_value, p_value = pearsonr(X, Y)
```

```python
print(f'Correlation Value : {r_value}')
print(f'P Value : {p_value}')
```

```python
import matplotlib.pyplot as plt

X = data[['Wash_Freq of BB (# times per week)']]
y = data['NumPhylum']


from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)


plt.figure(figsize=(10,6))


plt.scatter(X, y, color='blue', alpha=0.6, label='Observed Data')


plt.plot(X, y_pred, color='red', linewidth=2, label='Best-Fit Line')


plt.xlabel('Wash Frequency (times per week)')
plt.ylabel('Number of Phyla')
plt.title('Linear Regression: Wash Frequency vs Phylum Count')


plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Evaluation outputs
r_squared = model.score(X, y)
print(f"R (coefficient of determination): {r_squared:.3f}")
print(f"Intercept: {model.intercept_:.2f}")
print(f"Slope: {model.coef_[0]:.2f}")
```
